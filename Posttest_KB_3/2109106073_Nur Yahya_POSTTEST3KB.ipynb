```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("all_perth_house.csv")

sns.heatmap(df.corr(), cmap='flare', annot=True)
```

# HEATMAP
Berdasarkan heatmap diatas. ada banyak label yang memiliki kolerasi rendah, dengan nilai korelasi tertinggi berada 0.63 dan -0.64 dimana rumah dengan jarak stasiun terdekat memiliki korelasi positif dengan jarak rumah dengan sekolah terdekat. Dan latitude berkorelasi negatif dengan postcode

```python
st_df = df.sort_values(by='DATE_SOLD', ascending=True)
sns.lineplot(st_df.iloc[:100], x='DATE_SOLD', y='PRICE')
plt.xticks(rotation=90)
plt.xlabel("TANGGAL TERJUAL")
plt.ylabel("HARGA")
plt.title("DIAGRAM GARIS HARGA PENJUALAN RUMAH PERTAHUN")
plt.show()
```

## LINE PLOT
Berdasarkan line plot diatas, harga rumah di kota Perth, Australia mengalami kenaikan dari tahun 1990 sampai dengan 2010.
Tercatat harga tertinggi terjadi pada kisaran tahun 2007 dan mulai mengalami kenaikan yang lumayan stabil pada tahun 2002 setelah mengalami penurunan di tahun 2001.

```python
bedrooms = df['BEDROOMS'].value_counts()

plt.pie(bedrooms, labels = bedrooms.index, startangle=90, autopct = '%.2f%%')
plt.title("Number Of Bedrooms")
plt.show()
```

## PIE CHART
Berdasarkan pie chart diatas, mayoritas rumah yang berada di kota Perth Australia memiliki jumlah kamar sebanyak 4 dan 3. Lebih dari setengah jumlah keseluruhan rumah disana memiliki kamar sebanyak 4.

```python
bathrooms = df['BATHROOMS'].value_counts()
bathrooms.plot.bar()
plt.xticks(rotation=0)
plt.title("DATA KAMAR MANDI PADA TIAP RUMAH DI PERTH, AUSTRALIA")
plt.xlabel("BATHROOMS")
plt.ylabel("COUNT HOUSES")
plt.show()
```

## DIAGRAM BATANG
berdasarkan diagram batang diatas, mayoritas rumah yang berada di kota Perth, Australia memiliki 2 kamar mandi. Sebanyak lebih dari 20000 rumah di kota Perth, Australia yang memiliki 2 kamar mandi.
